```python
import io
import os
import sys
import types
import numpy as np
import pandas as pd
import altair as alt
import dropbox
```

```python
import knmi_gn_001 as gethdd
```

```python
alt.data_transformers.disable_max_rows()
```

```python
def regendata():
    
    df = gethdd.retrieveHDD()
    df['doy'] = df.index.to_series().dt.dayofyear
    df['yr'] = df.index.to_series().dt.year
    df['month'] = df.index.to_series().dt.month
    df['day'] = df.index.to_series().dt.day
    
    df['year'] = 2024
    df['normaldate'] =pd.to_datetime(df[['year','month', 'day']],format='%Y-%m-%d')
    
    df['regen_mm'] = df['RH']/10
    df['cumul_mm'] = df['regen_mm'].groupby(df['yr']).cumsum()
    
    df['RollWkSum_mm'] = df['regen_mm'].rolling(7).sum()  
    
    df['Weekbudget_mm'] = 11-df['RollWkSum_mm']
    df['Weekbudget_mm'] = df['Weekbudget_mm'].clip(0,1e6)
    
    return df
```

```python
df = regendata()
```

```python
df.head()
```

```python
def plotcumuls(df,firstyr=None):
    if not firstyr:
        firstyr = 2014
    source = df[df['yr']>=firstyr]
    
    lastyr = df['yr'].max()
    
    
    chart=alt.Chart(source[source['yr']!=lastyr].reset_index(),
                    title="Cumulative rain in Zwolle").mark_line(strokeWidth=1).encode(
        x='normaldate:T',
        y=alt.Y("cumul_mm:Q",
              axis=alt.Axis(title="Regen [mm]")),
        color='yr:N'
    ).properties(width=400,height=400).interactive()
    
    
    thisyr = alt.Chart(source[source['yr']==lastyr].reset_index()).mark_line(strokeWidth=3).encode(
        x='normaldate:T',
        y=alt.Y("cumul_mm:Q"),
        color='yr:N'
    )
    
    display(chart + thisyr)
    
    
    
```

```python
plotcumuls(df)
```
