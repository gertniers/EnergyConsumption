Q = Globale straling (in J/cm2) per uurvak / Global radiation (in J/cm2) during the hourly division
SQ = Duur van de zonneschijn (in 0.1 uren) per uurvak, berekend uit globale straling (-1 for <0.05 uur) / Sunshine duration (in 0.1 hour) during the hourly division, calculated from global radiation (-1 for <0.05 hour)
DR = Duur van de neerslag (in 0.1 uur) per uurvak / Precipitation duration (in 0.1 hour) during the hourly division
RH = Uursom van de neerslag (in 0.1 mm) (-1 voor <0.05 mm) / Hourly precipitation amount (in 0.1 mm) (-1 for <0.05 mm) 

```python
def RetrieveStation_r2_uurdata(station,period):
    
#     url = "https://cdn.knmi.nl/knmi/map/page/klimatologie/gegevens/daggegevens/etmgeg_278.zip"
    url1 = "https://cdn.knmi.nl/knmi/map/page/klimatologie/gegevens/uurgegevens/"
    url2 = "uurgeg_"
    url3 = ''.join([str(station),period])
    url4 = ".zip"
    url = [url1,url2,url3,url4]
    url = ''.join(url)
    
    chunk_size=128
    save_path = ''.join([url2,url3,url4])
    r = requests.get(url, stream=True)
    with open(save_path, 'wb') as fd:
        for chunk in r.iter_content(chunk_size=chunk_size):
            fd.write(chunk)
    
    with ZipFile(save_path) as z:
        fn_text = ''.join([url2,url3,".txt"])
        with z.open(fn_text) as f:
            df = pd.read_csv(f, header=1, delimiter=",",skiprows=29)
    
    df['YYYYMMDD'] = pd.to_datetime(df['YYYYMMDD'],format='%Y%m%d')
#     # remove leading white spaces in column names
    df.columns = df.columns.str.replace(' ','')
    df['tijd'] = pd.to_datetime(df['YYYYMMDD'].astype(str) +' '+ (df['HH']-1).astype(str) +':00:00')
    df['month_nr'] = df['tijd'].dt.month
    df['yr'] = df['tijd'].dt.year
    
    df.rename(columns={"YYYYMMDD": "dag"},inplace=True)
    df.set_index('tijd',inplace = True)
    cols = df.columns
    
#     df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
    
    return df
#     return save_path
```
