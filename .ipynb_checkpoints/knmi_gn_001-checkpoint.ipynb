```python
def chunk_splitter(raw):
    """
    Generator to read a raw file and yield chunks that are separated by 'empty lines': "# "
    Parameters
    ----------
    raw : str
    Yields
    -------
    str
    """
    chunk = []
    for line in raw.splitlines():
        if line == "# ":
            if len(chunk) == 0:
                continue
            else:
                yield chunk
                chunk = []
        else:
            chunk.append(line)
    else:
        yield chunk
```

```python
def RetrieveStation(stn):
    url = "http://projects.knmi.nl/klimatologie/daggegevens/getdata_dag.cgi"
    stations=[stn]
    params = { "stns": ":".join(str(station) for station in stations)}
    
    start='20180101'
    params.update({"start": start})

    variables = ['ALL']
    params.update({"vars": ":".join(variables)})
    
    r = requests.post(url=url, data=params)
    r.raise_for_status()
    
    chunks = chunk_splitter(raw=r.text)
    disclaimer = next(chunks)
    disclaimer = "\n".join(line.strip("# ") for line in disclaimer)
    stations_raw = next(chunks)
    
    # parse the legend
    legend_raw = next(chunks)
    # strip prefix '# ' and suffix '; '
    legend_raw = [entry.strip("# ").strip("; ") for entry in legend_raw]
    legend = {}
    for entry in legend_raw:
        sp = entry.split("=")
        # the key is the term before the first '='
        key = sp[0].strip()
        # the value is everthing that follows, so we rejoin everything with a '='
        value = "=".join(sp[1:]).strip()
        legend.update({key: value})
        
     # parse the header
    header = next(chunks)
    header = header[0]  # the header is only one line
    header = header.strip('# ').replace(' ', '')
    
#     parse the data
    data = next(chunks)
    lines = []
    for line in data:
        lines.append(line.strip('# ').replace(' ', ''))

    # join data and header
    lines.insert(0, header)
    data = "\n".join(lines)
    
    df = pd.read_csv(StringIO(data), index_col=1, converters={'YYYYMMDD': pd.Timestamp})
    df.index = pd.DatetimeIndex(df.index)
    return r, disclaimer, legend, header, df;
```

```python
def updateknmi(params_in):
    fn = "knmi_data.csv"
    doUpdate = False
    try:
        print("read the csv file with knmi data ",fn)
        df = pd.read_csv(fn,
                       parse_dates=['datetime'],
                       index_col = ['datetime'])
        a = max(df.index)
        print("last data point in csv file: ",a)

    except:
        print("File not found, start updating and processing raw data")
        a = datetime.fromisoformat('2000-01-01')
        doUpdate = True
        df = []

    # if last datapoint is older than X hours start updating the processed file
    b = datetime.now()
    print("The current time is :",b)
    difference = math.floor((b-a).total_seconds()/(3600*24))
    print("The number of days since last datapoint is :", difference)
    if difference > 0: doUpdate = True
    if doUpdate:
        # read the raw data
        df_ = GetKNMI(params_in)
        
        if not df_.empty:
            # Compare file with data from knmi. 
            
            
            df = df_ .copy()
            df.rename_axis('datetime',inplace = True)
            dowrite=df.to_csv(fn)
            print("data saved to file : ", fn)   
        else:
            print("No new data to append")
            print("")
        
    return(df)
```

```python
def retrieveHDD():
    cats = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
        
    df = updateknmi_r1('SQ,Q,DR,TG,FG,RH')
    # add rows to current date
    
    if df.empty == False :
        a=max(df.index)
        print("KNMI last datapoint is :",a)
        b = datetime.now()
        print("Current time :",b)
        difference = math.floor((b-a).total_seconds()/3600)
        print("Elapsed time since  [hours] : ",difference)
        NrRows2Add  = math.floor(difference / 24)
        print("Nr rows to add : ", NrRows2Add)

        datetime_series = pd.Series(pd.date_range(start=max(df.index)+pd.Timedelta(hours = 24)
                                              , end= datetime.now(), freq="D"))
        
        df_ = pd.DataFrame(index=datetime_series, columns=df.columns.tolist())
        df_ = df_.fillna(0) # with 0s rather than NaNs

        df = df.append(df_)

        df['HDD'] = (14-(df['TG']-2/3*df['FG'])/10).clip(0,10000)
        df = df.reset_index()
        df.columns.values[0] = ('datetime')
        df = df.set_index('datetime')
        df['dow'] = pd.Categorical(df.index.strftime('%A'))
        cat_type = CategoricalDtype(categories=cats, ordered=True)
        df['dow'] = df['dow'].astype(cat_type)
        print(df.tail())
    else:
        print("No KNMI data available")
        
    
    return(df)
```

```python
def RetrieveStation_r1(station):
    
#     url = "https://cdn.knmi.nl/knmi/map/page/klimatologie/gegevens/daggegevens/etmgeg_278.zip"
    url1 = "https://cdn.knmi.nl/knmi/map/page/klimatologie/gegevens/daggegevens/"
    url2 = "etmgeg_"
    url3 = str(station)
    url4 = ".zip"
    url = [url1,url2,url3,url4]
    url = ''.join(url)
    
    chunk_size=128
    save_path = ''.join([url2,url3,url4])
    r = requests.get(url, stream=True)
    with open(save_path, 'wb') as fd:
        for chunk in r.iter_content(chunk_size=chunk_size):
            fd.write(chunk)
    
    with ZipFile(save_path) as z:
        fn_text = ''.join([url2,url3,".txt"])
        with z.open(fn_text) as f:
            df = pd.read_csv(f, header=1, delimiter=",",skiprows=49)
    
    df['YYYYMMDD'] = pd.to_datetime(df['YYYYMMDD'],format='%Y%m%d')
    # remove leading white spaces in column names
    df.columns = df.columns.str.replace(' ','')
    df.rename(columns={"YYYYMMDD": "datetime"},inplace=True)
    df.set_index('datetime',inplace = True)
    cols = df.columns
    df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
    
    return df
#     return save_path
```

```python
def GetKNMI_r1(params_in):
    params = params_in.split(",")
    df = RetrieveStation_r1('278') #278 is Heino
    #response, disclaimer, legend, header, df = RetrieveStation('260') #260 is De Bilt
    
    return df[params]
```
