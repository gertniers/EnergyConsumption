Q = Globale straling (in J/cm2) per uurvak / Global radiation (in J/cm2) during the hourly division  
SQ = Duur van de zonneschijn (in 0.1 uren) per uurvak, berekend uit globale straling (-1 for <0.05 uur) / Sunshine duration (in 0.1 hour) during the hourly division, calculated from global radiation (-1 for <0.05 hour)   
DR = Duur van de neerslag (in 0.1 uur) per uurvak / Precipitation duration (in 0.1 hour) during the hourly division  
RH = Uursom van de neerslag (in 0.1 mm) (-1 voor <0.05 mm) / Hourly precipitation amount (in 0.1 mm) (-1 for <0.05 mm)

```python
def RetrieveStation_r2_uurdata(station,period):
    
#     url = "https://cdn.knmi.nl/knmi/map/page/klimatologie/gegevens/daggegevens/etmgeg_278.zip"
    url1 = "https://cdn.knmi.nl/knmi/map/page/klimatologie/gegevens/uurgegevens/"
    url2 = "uurgeg_"
    url3 = ''.join([str(station),period])
    url4 = ".zip"
    url = [url1,url2,url3,url4]
    url = ''.join(url)
    
    chunk_size=128
    save_path = ''.join([url2,url3,url4])
    r = requests.get(url, stream=True)
    with open(save_path, 'wb') as fd:
        for chunk in r.iter_content(chunk_size=chunk_size):
            fd.write(chunk)
    
    with ZipFile(save_path) as z:
        fn_text = ''.join([url2,url3,".txt"])
        with z.open(fn_text) as f:
            df = pd.read_csv(f, header=1, delimiter=",",skiprows=29)
    
    df['YYYYMMDD'] = pd.to_datetime(df['YYYYMMDD'],format='%Y%m%d')
#     # remove leading white spaces in column names
    df.columns = df.columns.str.replace(' ','')
    df['tijd'] = pd.to_datetime(df['YYYYMMDD'].astype(str) +' '+ (df['HH']-1).astype(str) +':00:00')
    df['month_nr'] = df['tijd'].dt.month
    df['yr'] = df['tijd'].dt.year
    
    df.rename(columns={"YYYYMMDD": "dag"},inplace=True)
    df.set_index('tijd',inplace = True)
    cols = df.columns
    
#     df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
    
    return df
#     return save_path
```

```python
def RetrieveStation_r1_uurdata(station):
    
#     url = "https://cdn.knmi.nl/knmi/map/page/klimatologie/gegevens/daggegevens/etmgeg_278.zip"
    url1 = "https://cdn.knmi.nl/knmi/map/page/klimatologie/gegevens/uurgegevens/"
    url2 = "uurgeg_"
    url3 = ''.join([str(station),'_2021-2030'])
    url4 = ".zip"
    url = [url1,url2,url3,url4]
    url = ''.join(url)
    
    chunk_size=128
    save_path = ''.join([url2,url3,url4])
    r = requests.get(url, stream=True)
    with open(save_path, 'wb') as fd:
        for chunk in r.iter_content(chunk_size=chunk_size):
            fd.write(chunk)
    
    with ZipFile(save_path) as z:
        fn_text = ''.join([url2,url3,".txt"])
        with z.open(fn_text) as f:
            df = pd.read_csv(f, header=1, delimiter=",",skiprows=29)
    
    df['YYYYMMDD'] = pd.to_datetime(df['YYYYMMDD'],format='%Y%m%d')
#     # remove leading white spaces in column names
    df.columns = df.columns.str.replace(' ','')
    df['tijd'] = pd.to_datetime(df['YYYYMMDD'].astype(str) +' '+ (df['HH']-1).astype(str) +':00:00')
    
    df.rename(columns={"YYYYMMDD": "dag"},inplace=True)
    df.set_index('tijd',inplace = True)
    cols = df.columns
    
#     df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
    
    return df
#     return save_path
```

```python
def GetKNMI_uurdata_r1(params_in):
    params = params_in.split(",")
    
    df1 = RetrieveStation_r2_uurdata(278,'_2011-2020') #278 is Heino
    df2 = RetrieveStation_r2_uurdata(278,'_2021-2030') #278 is Heino
    df = pd.concat([df1,df2])
    #response, disclaimer, legend, header, df = RetrieveStation('260') #260 is De Bilt
    
    return df[params]
```

# Opbrengst per m2 source
https://www.vattenfall.nl/kennis/opbrengst-zonnepanelen-per-m2/

1.78 m2  , 375 Wattpiek, 319 kWh/yr, 178 kWh/m2 yr  
2.0 m2,   400 Wp, 344 kWh/yr, 172 kWh/m2 yr   , 500 euro ex btw/paneel = 250 euro/m2 ex btw
1,822 m2,   310 Wp, 310 kWh / yr, 170 kWh/ m² yr. 490 euro ex btw/paneel = 272 euro/m2 ex btw.


https://uwduurzameinstallateur.nl/diensten/verduurzamen/zonnepanelen/opbrengst-zonnepanelen-per-m2/ 
anneer de panelen ideaal geplaatst zijn, zal een standaard zonnepaneel van 1,00 bij 1,65 meter van 250 wp gemiddeld ongeveer 215 kWh per jaar aan energie opwekken. Een vierkante meter zonnepaneel brengt dan gemiddeld ongeveer 120 kWh aan energie op per jaar 



Kies efficiency zo, dat jaaropbrengst voor 10 m2 gelijk is aan 1700 kWh/yr 


Andere source: vermenigvuld Wattpiek met 0.85 om tot gemiddeld jaaropbrengst in kWh te komen:
170 * 0.85 = 145 kWh/m2.   


Andere source: 200 Wp / m2 geeft 177 kWh/m2 yr (0.885 factor ipv 0.85)  

10 m2 met 200 Wp/ m2 --> 1770 kWh/m2 per jaar

```python
def pick_sample(df,inputdate):
    ''' select random a day from dataset with consumption
    '''
    # mnth = datetime.strptime(inputdate, '%Y-%M-%d').month
    mnth = inputdate.month
    kwart = inputdate.quarter
    dow = inputdate.dayofweek
    weekend = math.floor(dow/5)
    
    
    mask = (df['kwart']==kwart) & (df['weekend']==weekend) & (df['datetime'].dt.year>=2021)
    
    trekking = df[mask].sample(1)   
    return trekking['dy'].squeeze()
```

```python
def bereken_opbrengst(df,Wp):
    df = df.copy()
    ''' bereken eerst totalen van zonnneopbrengst
    verdeel later naar laag en hoog. 
    
    '''

    df['opbrengst'] = df['Q_kWh_perWp']*Wp
    df['gebruik_eigen'] = df.apply(lambda x: min(x['Elektriciteit'],x['opbrengst']),axis=1)
    df['gebruik_net'] = df.apply(lambda x: max(0,(x['Elektriciteit']-x['opbrengst'])),axis=1)
    df['terugleveren'] = df.apply(lambda x: max(0,(x['opbrengst']-x['Elektriciteit'])),axis=1)                      
   
    cols = ['opbrengst','gebruik_eigen','gebruik_net','terugleveren']
    
    
    # verdeel naar laag en hoog tarief
    df['zon_laag_kWh']=df.apply(lambda x: x['opbrengst'] * x['LaagTariefSwitch'],axis=1)
    df['zon_hoog_kWh']=df.apply(lambda x: x['opbrengst'] * (1-x['LaagTariefSwitch']),axis=1)
    df['eigengebruik_laag_kWh']=df.apply(lambda x: x['gebruik_eigen'] * x['LaagTariefSwitch'],axis=1)
    df['eigengebruik_hoog_kWh']=df.apply(lambda x: x['gebruik_eigen'] * (1-x['LaagTariefSwitch']),axis=1)
    df['gebruiknet_laag_kWh'] = df.apply(lambda x: x['gebruik_net'] * x['LaagTariefSwitch'],axis=1)
    df['gebruiknet_hoog_kWh'] = df.apply(lambda x: x['gebruik_net'] * (1-x['LaagTariefSwitch']),axis=1)
    df['teruglever_laag_kWh'] = df.apply(lambda x: x['terugleveren'] * x['LaagTariefSwitch'],axis=1)
    df['teruglever_hoog_kWh'] = df.apply(lambda x: x['terugleveren'] * (1-x['LaagTariefSwitch']),axis=1)
    
 
    df['Wp'] = Wp
    # return df[cols]
    return df
```
